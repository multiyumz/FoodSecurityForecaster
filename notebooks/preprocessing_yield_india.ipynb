```python

import numpy as np
import pandas as pd
```

# Import y values

```python
india = pd.read_csv('../raw_data/Crop yield data/COUNTY_level_annual/india.csv')
india_cleaned = india[india['YIELD.RM.OUT'].isna() == False].drop(columns=['SEASON', 'Unnamed: 0', 'YIELD.RM.OUT'])
india_cleaned.head(4)
```

```python
# largest crops
india_cleaned.groupby('CROP').sum().sort_values(by='AREA', ascending=False).head(5)
```

```python
india_rice = india_cleaned[ india_cleaned['CROP'] == 'RICE' ]
```

```python
india_rice = india_rice.groupby(['STATE', 'YEAR']).mean().reset_index()
india_rice = india_rice.sort_values(by=['STATE', 'YEAR'])
india_rice = india_rice [ india_rice['YEAR'] >= 2000]
india_rice.head(7)
```

```python
india_rice_csv = india_rice.to_csv('india_rice.csv')
```

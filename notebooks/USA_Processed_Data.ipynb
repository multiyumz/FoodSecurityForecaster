```python
data = pd.read_csv("../raw_data/Crop yield data/COUNTY_level_annual/USA.csv")
```

```python
data = data.drop(columns=['Data Item','Week Ending'])
```

```python
data = data.rename(columns={"Year": "YEAR", "State": "STATE","Commodity":"COMMODITY","Yield":"YIELD"})
```

```python
data.groupby('COMMODITY').sum().sort_values(by='YIELD', ascending=False)
```

```python
soybeans_usa = data[data['COMMODITY'] == 'SOYBEANS']
```

```python
soybeans_usa = soybeans_usa.sort_values(by=['STATE', 'YEAR']).reset_index().drop(columns='index')
```
